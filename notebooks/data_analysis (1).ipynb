```python
import pandas as pd

df = pd.read_csv(
    r'C:\Users\Dell\Downloads\city_day_cleaned.csv',
    parse_dates=['date']
)
df.head()
```

```python
# feature Engineering

df["date"] = pd.to_datetime(df["date"])
```

```python
df["is_weekend"] = df["date"].dt.weekday.isin([5, 6]).astype(int)
df["is_weekend"].value_counts()
```

```python
def set_day_time(hour: int) -> int:
    if 6 <= hour <= 9:
        return 0  # morning
    elif 9 < hour <= 12:
        return 1  # midday
    elif 12 < hour <= 17:
        return 2  # afternoon
    elif 17 < hour <= 21:
        return 3  # evening
    else:
        return 4  # night

df["day_time"] = df["date"].dt.hour.apply(set_day_time)
df["day_time"].value_counts()
```

```python
df = df.drop(columns=["day_time"])
```

```python
def set_season(month: int):
    if month in [12, 1, 2]:
        return 0  # Winter
    elif month in [3, 4, 5]:
        return 1  # Summer
    elif month in [6, 7, 8, 9]:
        return 2  # Monsoon
    else:
        return 3  # Autumn

df["season"] = df["date"].dt.month.apply(set_season)
df["season"].value_counts()
```

```python
df = df.drop(columns=["date"])
```

```python
df = pd.read_csv("../data/preprocessed/preprocessed.csv")
```

```python
X = df.drop("aqi", axis=1)
y = df["aqi"]

print(X.shape, y.shape)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, shuffle=True
)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.model_selection import cross_val_score, KFold
import numpy as np

models = {
    "Linear Regression": LinearRegression(),
    "Random Forest": RandomForestRegressor(n_estimators=50, max_depth=10, n_jobs=-1),
    "Gradient Boosting": GradientBoostingRegressor(n_estimators=50)
}

cv = KFold(n_splits=5, shuffle=True, random_state=42)

results = []

for name, model in models.items():
    rmse = np.sqrt(
        -cross_val_score(
            model, X_train, y_train,
            scoring="neg_mean_squared_error",
            cv=cv
        ).mean()
    )
    r2 = cross_val_score(model, X_train, y_train, scoring="r2", cv=cv).mean()
    
    results.append((name, rmse, r2))

results
```

```python
from sklearn.model_selection import GridSearchCV

param_grid = {
    "n_estimators": [100, 200],
    "max_depth": [10, 20, None],
    "min_samples_split": [2, 5],
    "min_samples_leaf": [1, 2]
}

rf = RandomForestRegressor(random_state=42, n_jobs=-1)

grid = GridSearchCV(
    rf,
    param_grid,
    cv=3,
    scoring="neg_mean_squared_error",
    n_jobs=-1
)

grid.fit(X_train, y_train)

best_model = grid.best_estimator_
print(grid.best_params_)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import numpy as np

y_pred = best_model.predict(X_test)

rmse = np.sqrt(mean_squared_error(y_test, y_pred))
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("Final RMSE:", rmse)
print("Final MAE:", mae)
print("Final R²:", r2)
```

```python
import pandas as pd

feature_importance = pd.Series(
    best_model.feature_importances_,
    index=X.columns
).sort_values(ascending=False)

print(feature_importance.head(10))
```

```python
residuals = y_test - y_pred

print("Residual mean:", residuals.mean())
print("Residual std:", residuals.std())
```

```python
y_train_pred = best_model.predict(X_train)

train_rmse = np.sqrt(mean_squared_error(y_train, y_train_pred))
test_rmse = rmse

print("Train RMSE:", train_rmse)
print("Test RMSE:", test_rmse)
```
